```python
#task is to classify  churn (or  if customer close accounts in bank based on  several criterion)
import pandas as pd
#import  all ensemble  algorithms
from sklearn.ensemble import VotingClassifier,StackingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
scaler =MinMaxScaler()
bank =pd.read_csv("churn_classification.csv")
bank['gender'] =bank['gender'].map({"Female":0,"Male":1})
y =bank['churn'].values
X =bank.drop('churn',axis=1).values
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2,random_state=1)
X_train =scaler.fit_transform(X_train)
X_test =scaler.transform(X_test)
#use voting classifier  based on hard voting principle
hybrid =VotingClassifier(
    estimators =[('SVM',SVC(probability=True)),('gauss',GaussianNB()),('tree',DecisionTreeClassifier()),
                 ('neighbour',KNeighborsClassifier())],
    voting='soft'
)
hybrid.fit(X_train,y_train)
print(hybrid.score(X_test,y_test))

#creating of stacking model
stacking_model =StackingClassifier(
    estimators =[('SVM',SVC(probability=True)),('gauss',GaussianNB()),('tree',DecisionTreeClassifier()),
                 ('neighbour',KNeighborsClassifier())],final_estimator =LogisticRegression(max_iter=67000)
)

stacking_model.fit(X_train,y_train)
print(stacking_model.score(X_test,y_test))
bank.head()
```

```python
for name, model in hybrid.named_estimators_.items():
  print(f'score of {name} is given as {model.score(X_test,y_test)}')
```

```python
#now we are using gridsearchcv strategy to find optimal parameters for given algorithms
from sklearn.model_selection import GridSearchCV
parameters ={
    'neighbour__n_neighbors':[3,5,7,9,11,13,15]
}
mymodel =GridSearchCV(estimator =hybrid,param_grid =parameters, cv=3,scoring='accuracy',n_jobs=-1)
mymodel.fit(X_train,y_train)
print(mymodel.best_score_)
```

```python
# Implementing Neural Network with 7 layers, 40 nodes each
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create neural network with 7 layers, 40 nodes each
nn_model = Sequential([
    Dense(40, activation='relu', input_shape=(X_train.shape[1],)),  # Layer 1
    Dense(40, activation='relu'),  # Layer 2
    Dense(40, activation='relu'),  # Layer 3
    Dense(40, activation='relu'),  # Layer 4
    Dense(40, activation='relu'),  # Layer 5
    Dense(40, activation='relu'),  # Layer 6
    Dense(40, activation='relu'),  # Layer 7
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile the model
nn_model.compile(optimizer=Adam(learning_rate=0.001),
                 loss='binary_crossentropy',
                 metrics=['accuracy'])

# Display model architecture
nn_model.summary()
```
