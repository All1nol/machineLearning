```python
#now let us see structure of  folder and read images
import cv2  # this is for  computer vision
import os  #this for folder structure or operating system command
subdirectories =os.listdir("train") #create list
subdirectories.sort()
print(subdirectories)
X =[]  #save images
y =[]  #save  labels
for label, name in enumerate(subdirectories):
  sub_folder =os.path.join("train",name)
  # print(label, sub_folder)
  for file in os.listdir(sub_folder): #another loop for accessing file
      image =cv2.imread(os.path.join(sub_folder,file))  #actual image reading
    #print(os.path.join(sub_folder,file))
      image =cv2.resize(image,(128,128)) #all images has this size
      X.append(image)  #put images into array
      y.append(label)  #put label into array

```

```python
#convert all images to the tensor or numpy matrix
import numpy as np
X =np.array(X)
y =np.array(y)
```

```python
X.shape
```

```python
#divide data into training and test part
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.15,random_state=1)
```

```python
#scale images such thata their value should between [0,1] it means
#we have to divide it by 255
X_train =X_train.astype('float32')  /255.0
X_test =X_test.astype('float32')  /255.0
```

```python
#now create model
#we  have to use  pretrained  or already existed model
from keras.applications import VGG19
#VGG19 is  pretrained or someone else trained  for some problem
# and i have to use it for my problem
#this is called as  transfer learning
base =VGG19(
    include_top =False,input_shape=(128,128,3)
)
#shape (300,300,3) coincides  to cv2.resize(image,(300,300))
# and  3  denotes  number of channel or RGB -red green blue
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten
#now we are creating normal neural network
model =Sequential()
model.add(base)  #add  pretrained model
model.add(Flatten())  #convert image to vector
model.add(Dense(units=100,activation='relu'))
model.add(Dense(units=50,activation='relu'))
model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=10,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(optimizer='sgd',loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(X_train,y_train,batch_size=50,epochs=50,
          validation_data =(X_test,y_test),verbose=1)
```

```python
test =cv2.imread("/content/person520_bacteria_2205.jpeg")
test =cv2.resize(test,(128,128))
test =np.expand_dims(test,axis=0)
print(model.predict(test))
#if it is less  than 0.5, it is class 0, else class 1
#because class 1 is pheumonia and result is [[1.]]
#that means it is image of pheumonia
test.shape
```
